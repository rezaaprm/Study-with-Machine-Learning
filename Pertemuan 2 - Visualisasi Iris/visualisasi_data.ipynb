# Visualisasi Data - Iris Dataset

```python
# lib manipulasi data
import pandas as pd
import numpy as np

# lib visualisasi data
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("../Dataset/iris.csv")
dataset.info()
```

```python
dataset
```

```python
# Menampilkan jumlah data
df = dataset["species"].value_counts().reset_index()
print(df)
```

```python
# Membuat bingkai
fig, ax = plt.subplots(figsize=(8,4))

# Membuat barplot
plt.bar(x=df["species"], height=df["count"])

# Set labels
ax.set_title("Barplot Iris Flowers", fontsize=14)
ax.set_xlabel("Species", fontsize=12)
ax.set_ylabel("Jumlah Data", fontsize=12)
ax.grid(True)

# Menampilkan data
plt.show()
```

```python
# Membuat bingkai
fig, ax = plt.subplots(figsize=(8,4))

# Membuat barplot seaborn
sns.barplot(data=df, x="species", y="count", hue="species", legend=True)

# Set labels
ax.set_title("Barplot Iris Flowers", fontsize=14)
ax.set_xlabel("Species", fontsize=12)
ax.set_ylabel("Jumlah Data", fontsize=12)
ax.grid(True)

# Menampilkan data
plt.show()
```

```python
# Membuat bingkai
fig, ax = plt.subplots(figsize=(8,4))

# Membuat barplot plt
plt.pie(x=df["count"], labels=df["species"], autopct='%1.1f%%')

# Set labels
# ax.set_title("Barplot Iris Flowers", fontsize=14)
# ax.set_xlabel("Species", fontsize=12)
# ax.set_ylabel("Jumlah Data", fontsize=12)
# ax.grid(True)

# Menampilkan data
plt.show()
```

```python
# Membuat bingkai
fig, ax = plt.subplots(figsize=(8,4))

# Membuat barplot seaborn
plt.pie(x=df["count"], labels=df["species"], autopct='%1.1f%%', explode=(0.2, 0.00, 0.00), startangle=360)

# Set labels
# ax.set_title("Barplot Iris Flowers", fontsize=14)
# ax.set_xlabel("Species", fontsize=12)
# ax.set_ylabel("Jumlah Data", fontsize=12)
# ax.grid(True)

# Menampilkan data
plt.show()
```

```python
df_corr = dataset.corr(numeric_only=True, method="pearson")
```

```python
df_corr
```

```python
# Membuat bingkai
fig, ax = plt.subplots(figsize=(8,4))

# Membuat heatmap seaborn
sns.heatmap(data=df_corr, vmin=-1, vmax=1, cmap="magma", annot=True, fmt='.3g', linewidths=0.5, linecolor="white")

# Set labels
# ax.set_title("Barplot Iris Flowers", fontsize=14)
# ax.set_xlabel("Species", fontsize=12)
# ax.set_ylabel("Jumlah Data", fontsize=12)
# ax.grid(True)

# Menampilkan data
plt.show()
```

```python
# Membuat bingkai
fig, ax = plt.subplots(figsize=(8,4))

# Membuat scatter seaborn
sns.scatterplot(data=dataset, x="petal_length", y="petal_width", hue="species")

# Set labels
ax.set_title("Scatterplot Iris Flowers", fontsize=14)
ax.set_xlabel("petal_length", fontsize=12)
ax.set_ylabel("petal_width", fontsize=12)
ax.grid(True)

# Menampilkan data
plt.show()
```

```python
# Membuat bingkai
fig, ax = plt.subplots(figsize=(8,4))

# Membuat boxplot matplotlib
plt.boxplot(x=dataset[["petal_length", "petal_width", "sepal_length", "sepal_width"]],
            labels=["petal_length", "petal_width", "sepal_length", "sepal_width"],
            patch_artist=True, widths=(0.5, 0.5, 0.5, 0.5))

# Set labels
ax.set_title("Boxplot Iris Flowers", fontsize=14)
ax.set_xlabel("", fontsize=12)
ax.set_ylabel("", fontsize=12)
ax.grid(True)

# Menampilkan data
plt.show()
```

```python
# Membuat bingkai
fig, ax = plt.subplots(figsize=(8,4))

# Membuat boxplot seaborn
sns.boxplot(data=dataset)

# Set labels
ax.set_title("Boxplot Iris Flowers", fontsize=14)
ax.set_xlabel("", fontsize=12)
ax.set_ylabel("", fontsize=12)
ax.grid(True)

# Menampilkan data
plt.show()
```
