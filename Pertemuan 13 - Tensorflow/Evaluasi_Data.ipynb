```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#create DataFrame
df = pd.DataFrame({'Batch Size': [8, 16, 24, 32, 8, 16, 24, 32, 8, 16, 24, 32, 8, 16, 24, 32],
                   'R': [0.9864, 0.9949, 0.9939, 0.9896,
                         0.997, 0.995, 0.9951, 0.9874,
                         0.9898, 0.9927, 0.994, 0.9853,
                         0.9938, 0.9933, 0.9946, 0.9825],
                   'MAPE': [16.54, 3.28, 11.47, 7.64,
                            4.20, 4.40, 7.78, 7.88,
                            13.81, 9.62, 7.47, 8.60,
                            13.41, 12.40, 7.08, 8.78],
                   'Optimizer': ['Adam', 'Adam', 'Adam', 'Adam',
                                 'Adamax', 'Adamax', 'Adamax', 'Adamax',
                                 'RMSprop', 'RMSprop', 'RMSprop', 'RMSprop',
                                 'SGD', 'SGD', 'SGD', 'SGD']
                  })

#view DataFrame
print(df)
```

```python
# Plot R values
fig, ax1 = plt.subplots(figsize=(8, 4))

#set seaborn plotting aesthetics
sns.set(style='white')

#create grouped bar chart for R
sns.barplot(x='Batch Size', y='R', hue='Optimizer', data=df, ax=ax1) 

# Labels for R plot
ax1.set_title('Evaluation Model using R with 25 Epochs', fontsize=14)
ax1.set_xlabel('Batch Size', fontsize=14)
ax1.set_ylabel('R', fontsize=14)
ax1.legend(loc="best", ncols=2)
ax1.grid(True)

plt.tight_layout()
plt.show()

# Plot MAPE values
fig, ax2 = plt.subplots(figsize=(8, 4))

#create grouped bar chart for MAPE
sns.barplot(x='Batch Size', y='MAPE', hue='Optimizer', data=df, ax=ax2) 

# Labels for MAPE plot
ax2.set_title('Evaluation Model using MAPE with 25 Epochs', fontsize=14)
ax2.set_xlabel('Batch Size', fontsize=14)
ax2.set_ylabel('MAPE (%)', fontsize=14)
ax2.legend(loc="best", ncols=2)
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#create DataFrame dengan Run Time baru
df = pd.DataFrame({'Batch Size': [8, 16, 24, 32, 8, 16, 24, 32, 8, 16, 24, 32, 8, 16, 24, 32,
                                  8, 16, 24, 32, 8, 16, 24, 32, 8, 16, 24, 32, 8, 16, 24, 32],
                   'Run Time': [363.08, 377.34, 358.6, 391.78,
                                176.32, 203.77, 206.89, 203.21,
                                150.42, 162.9, 147.36, 144.23,
                                153.8, 130.12, 152.65, 103.17,
                                772.08, 756.99, 689.22, 663.24,
                                414.94, 349.41, 401.37, 373.67,
                                294.2, 289.12, 284.04, 230.58,
                                255.33, 250.4, 246.02, 190.91],
                   'Optimizer': ['Adam', 'Adam', 'Adam', 'Adam',
                                 'Adamax', 'Adamax', 'Adamax', 'Adamax',
                                 'RMSprop', 'RMSprop', 'RMSprop', 'RMSprop',
                                 'SGD', 'SGD', 'SGD', 'SGD',
                                 'Adam', 'Adam', 'Adam', 'Adam',
                                 'Adamax', 'Adamax', 'Adamax', 'Adamax',
                                 'RMSprop', 'RMSprop', 'RMSprop', 'RMSprop',
                                 'SGD', 'SGD', 'SGD', 'SGD']
                  })

#view DataFrame
print(df)
```

```python
# Plot Run Time values
fig, ax = plt.subplots(figsize=(10, 6))

#set seaborn plotting aesthetics
sns.set(style='whitegrid')

#create grouped line chart
sns.lineplot(x='Batch Size', y='Run Time', hue='Optimizer', data=df, marker='o', ax=ax, alpha=0.7) 

# Customize labels and title
ax.set_title('Model Run Time', fontsize=16, fontweight='bold', color='darkblue')
ax.set_xlabel('Batch Size', fontsize=14, fontweight='bold')
ax.set_ylabel('Run Time (seconds)', fontsize=14, fontweight='bold')
ax.legend(loc="best", ncol=2, fontsize=12, title='Optimizer')
ax.grid(True, linestyle='--', alpha=0.7)

# Customize ticks
ax.tick_params(axis='both', which='major', labelsize=12)
ax.set_xticks([8, 16, 24, 32])
ax.set_yticks(range(100, 900, 100))

# Add data labels
for line in ax.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax.text(x, y, f'{y:.1f}', ha='center', va='bottom', fontsize=10, color=line.get_color())

plt.tight_layout()
plt.show()
```
