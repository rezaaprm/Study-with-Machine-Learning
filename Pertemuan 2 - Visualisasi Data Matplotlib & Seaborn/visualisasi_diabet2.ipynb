### Exploration Data Analysis - Pima Diabetes

```python
## Library Manipulation Data
import numpy as np
import pandas as pd

## Library Visualisasi Data
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
dataset = pd.read_csv("../dataset/dataset_diabetes.csv")
dataset.info()
```

```python
# Cek missing values
print(dataset.isnull().sum())
```

```python
# Cek missing values
print(dataset.isna().sum())
```

```python
# Display dataset
print(dataset)
```

```python
dataset
```

```python
# Hitung korelasi menggunakan metode pearson
df_corr = dataset.corr(method="pearson")

# create heatmap
fig, ax = plt.subplots(figsize=(10,5))
sns.heatmap(df_corr, vmin=-1, vmax=1, cmap="crest", annot=True, fmt='.3g', linecolor="white", linewidth=0.5)

# show plot
plt.show()
```

```python
# create barplot
fig, ax = plt.subplots(figsize=(8,4))
sns.scatterplot(data=dataset, x="Age", y="Glucose", hue="Outcome")

# set labels
ax.set_title("Scatterplot Age and Glucose", fontsize=14)
ax.set_xlabel("", fontsize=12)
ax.set_ylabel("", fontsize=12)
ax.grid(True)

# show plot
plt.show()
```

```python
# hitung jumlah datasetnya
df = dataset["Outcome"].value_counts().rename("Count").reset_index()
```

```python
# create barplot
fig, ax = plt.subplots(figsize=(8,4))
ax.pie(x=df["Count"], labels=df["Outcome"], autopct='%1.2f%%', explode=(0.01, 0.01))

# show plot
plt.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12,4))

# pieplot
ax[0].pie(x=df["Count"], labels=df["Outcome"], autopct='%1.2f%%', explode=(0.01, 0.01))
ax[0].grid(False)

# barplot
sns.barplot(data=df, x="Outcome", y="Count", hue="Outcome", ax=ax[1])
ax[1].grid(True)

# set labels
# ax.set_title("Scatterplot Age and Glucose", fontsize=14)
# ax.set_xlabel("", fontsize=12)
# ax.set_ylabel("", fontsize=12)
# ax.grid(True)

# show plot
plt.show()
```
