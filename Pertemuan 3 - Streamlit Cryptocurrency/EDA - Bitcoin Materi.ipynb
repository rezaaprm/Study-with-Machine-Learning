## EDA - Time Series Bitcoin Cryptocurrency

```python
## Library Manipulation Data
import numpy as np
import pandas as pd

## Library Visualisasi Data
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
dataset = pd.read_csv("../dataset/BTC-USD update 08-2024.csv", parse_dates=["Date"])
dataset.info()
```

```python
# Display Dataset
dataset
```

```python
# create frame
fig, ax = plt.subplots(figsize = (8,4))

# time series plot
ax.plot(dataset["Date"], dataset["Open"], color="tab:green", label="Open Price", linewidth=2)
ax.plot(dataset["Date"], dataset["High"], color="tab:orange", label="High Price", linewidth=2)
ax.plot(dataset["Date"], dataset["Low"], color="tab:red", label="Low Price", linewidth=2)
ax.plot(dataset["Date"], dataset["Close"], color="tab:blue", label="Close Price", linewidth=2)

# set label-labels
ax.set_title("Time Series Bitcoin Price",fontsize=12)
ax.set_xlabel("",fontsize=12)
ax.set_ylabel("", fontsize=12)
ax.legend(loc="best")
ax.grid(True)

# return values
plt.show()
```
