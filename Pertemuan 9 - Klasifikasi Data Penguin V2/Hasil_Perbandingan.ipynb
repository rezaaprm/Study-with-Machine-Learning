## Perbandingan Klasifikasi

```python
# Library Utama
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    "performa" : ["akurasi", "presisi", "recall", "f1-score"],
    "SVM_80_20" : [0.96, 0.94, 0.95, 0.94],
    "SVM_70_30" : [0.97, 0.96, 0.97, 0.96],
    "LR_80_20" : [0.97, 0.98, 0.94, 0.96],
    "LR_70_30" : [0.95, 0.97, 0.91, 0.93],
}
df = pd.DataFrame(data)
df.info()
```

```python
x = np.arange(4)
y1 = df["SVM_80_20"]
y2 = df["SVM_70_30"]
width = 0.4

plt.bar(x-0.2, y1, width, color="blue")
plt.bar(x+0.2, y2, width, color="red")

plt.xticks(x, ["akurasi", "presisi", "recall", "f1-score"])
plt.title("Klasifikasi Penguin menggunakan Metode Support Vector Machine", fontsize=14)
plt.xlabel("Performa", fontsize=12)
plt.ylabel("Nilai Performa", fontsize=12)
plt.tight_layout()

plt.legend(["80% & 20%", "70% & 30%"])
plt.show()
```

```python
x = np.arange(4)
y1 = df["LR_80_20"]
y2 = df["LR_70_30"]
width = 0.4

plt.bar(x-0.2, y1, width, color="blue")
plt.bar(x+0.2, y2, width, color="red")

plt.xticks(x, ["akurasi", "presisi", "recall", "f1-score"])
plt.title("Klasifikasi Penguin menggunakan Metode Logistic Regression", fontsize=14)
plt.xlabel("Performa", fontsize=12)
plt.ylabel("Nilai Performa", fontsize=12)
plt.tight_layout()

plt.legend(["80% & 20%", "70% & 30%"])
plt.show()
```
